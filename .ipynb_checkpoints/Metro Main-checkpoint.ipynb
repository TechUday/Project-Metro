```python
from sklearn.preprocessing import StandardScaler
```

```python
# Initialize StandardScaler
scaler = StandardScaler()
```

```python
data = np.array(Y).reshape(-1, 1)

# Fit the scaler to the data
scaler.fit(data)
```

```python
# Save the scaler as a pickle file
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
```

```python
Y_NEW = scaler.transform(data)
```

```python
Y_NEW
```

```python
Y_NEW = Y_NEW.reshape(-1)
```

```python
l2_X_train_n, l2_X_test_n, y_train_n, y_test_n = train_test_split(l2_normalized, Y_NEW, test_size=0.2, random_state=42)
```

```python
# create regressor object
regressor_l2_n = RandomForestRegressor(n_estimators=100,
                                  random_state=0)
 
# fit the regressor with x and y data
regressor_l2_n.fit(l2_X_train_n, y_train_n)
```

```python
y_pred_l2_n = regressor_l2_n.predict(l2_X_test_n)
```

```python
y_pred_l2_n_hypo = scaler.inverse_transform(y_pred_l2_n)
```

```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import sklearn
import pickle
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation,Dense

%matplotlib inline
```

```python
#initialising the ANN
model= Sequential()

#adding the input layer as 7 columns of inputs are there
model.add(Dense(units=64,activation='relu',input_shape=[55]))
#adding the second hidden layer
model.add(Dense(units=64,activation='relu'))
#adding the third hidden layer
model.add(Dense(units=64,activation='relu'))


#adding the output layer
model.add(Dense(units=1))
```

```python
#compiling the ANN
model.compile(loss='mean_squared_error',optimizer='Adam',metrics=['mean_squared_error','mean_absolute_error'])
```

```python
model.summary()
```

```python
model.fit(l2_X_train_n, y_train_n,batch_size=100,epochs=100)
```

```python
prediction=model.predict(l2_X_test_n)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
rmse = np.sqrt(mean_squared_error(y_test_n,prediction))
```
